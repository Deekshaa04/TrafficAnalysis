```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel('/content/transformed_data.xlsx')
```

```python
weather_impact = data.groupby('weather_condition').agg({'vehicle_count': 'mean', 'average_speed': 'mean'}).reset_index()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='weather_condition', y='vehicle_count', data=weather_impact)
plt.title('Average Vehicle Count by Weather Condition')
plt.show()
```
